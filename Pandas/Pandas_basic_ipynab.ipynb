### importing pandas

```python
import numpy as np
import pandas as pd
```

### series from list

```python
#series
country = ['india', 'pakistan', 'bangladesh']
pd.Series(country)
#Series is a class and country is sent as object
#dtype: object: every element dtype is objects. in most case, panda identify string as object
```

```python
#integer
runs = [100, 200, 300]
pd.Series(runs)
```

### custom index

```python
marks = [56, 77, 92]
subjects = ['maths', 'science', 'english']
pd.Series(marks, index = subjects)
pd.Series(marks, subjects) #same but shorter
```

### setting a name

```python
pd.Series(marks, index = subjects, name = 'makrs of ehtesam')
```
