### numpy array vs python list

### speed

```python
#speed
#list
a = [i for i in range(1000000)]
b = [i for i in range(1000000, 20000000)]
c = []

#measure time
import time

start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
end = time.time();
print(end-start)    #seconds
```

```python
#in numpy
import numpy as np

a = np.arange(10000000)
b = np.arange(10000000, 20000000)

start = time.time()
c = a + b
end = time.time();
print(end-start)
```

### brodcasting
Broadcasting is a way to perform operations on arrays of different shapes.

NumPy automatically “expands” the smaller array to match the bigger one.

### brodcasting rules

✅ Rule 1: Right-to-left comparison

Shapes are compared starting from the last (rightmost) dimension and moving left.
If one array has fewer dimensions, it is treated as if it has leading dimensions of size 1.

✅ Rule 2: Dimension compatibility

For each dimension, sizes are compatible only if:

They are equal, OR
One of them is 1

✅ Rule 3: Expansion of dimension 1

If a dimension is 1, it is stretched (broadcasted) to match the size of the other array in that dimension.
This expansion is conceptual (no actual data copying needed internally).

❌ Rule 4: Incompatibility leads to error

If in any dimension:

sizes are different and
neither is 1

→ broadcasting fails and NumPy raises an error.

### brodcasting example

```python
a = np.arange(12).reshape(3, 4)
b = np.arange(4).reshape(4)
print(a)
print(b)
```

```python
a + b
```

```python
#more example
a = np.arange(12).reshape(3, 4)
b = np.arange(4).reshape(1, 4)
print(a)
print(b)
```

```python
a = np.arange(12).reshape(3, 4)
b = np.arange(3)
print(a)
print(b)
print(a + b)
#error (3, 4) + (3, 3) brodcast not possible
```

```python
a = np.arange(3).reshape(1, 3)
b = np.arange(3).reshape(3, 1)
print(a)
print(b)
print(a + b)
#sucess : (3, 3) + (3, 3)..every 1 will scaled up to 3
```

```python
a = np.arange(4).reshape(1, 4)
b = np.arange(3).reshape(3, 1)
print(a)
print(b)
print(a + b) #(3, 4) + (3, 4)
```

```python
a = np.arange(1)
# first (1, 1)
#then (2, 2)
b = np.arange(4).reshape(2, 2)
print(a)
print(b)
print(a + b) #(2, 2) + (2, 2)
```

```python
a = np.arange(12).reshape(3, 4)
b = np.arange(12).reshape(4, 3)
print(a)
print(b)
print(a + b) #error
```

```python
a = np.arange(16).reshape(4, 4)
b = np.arange(4).reshape(2, 2)
print(a)
print(b)
print(a + b) #error
```

### sigmoid

sigmoid formula : σ(x) = 1 / (1 + e^(-x))

```python
#custom function to calculate sigmoid (all output is in range of 0 to 1 )
def sigmoid(arr):
  return 1/(1 + np.exp(-arr))
  #NumPy automatically applies operations element
  #So even though i passed an array, the function works on every element

a = np.arange(12)
sigmoid(a)
```

###loss function
A loss function measures how wrong your model is

### mean squared error

MSE = loss function for regression

Penalizes large errors more than small ones

Always non-negative

MSE = (1/n) * Σ (y_true - y_pred)²

```python
actual = np.random.randint(1, 50, 25)
predicted = np.random.randint(1, 50, 25)
print(actual)
print(predicted)
```

```python
def mse(acutal, predicted):
  return np.mean((actual - predicted) ** 2)

mse(actual, predicted)
```

### Categorical Cross-Entropy
Categorical Cross-Entropy = loss for multi-class classification

measures difference between true label and predicted probability

CCE = - Σ (y_true * log(y_pred))

```python
actual = np.array([1, 0, 0])
predicted = np.array([0.8, 0.1, 0.1])

def cce(actual, predicted):
  s = np.sum(actual * np.log(predicted))
  return -1 * s

cce(actual, predicted)
```

### binary cross entropy

BCE = loss function for binary classification

Works with sigmoid activation

Output must be probability (0–1)



bce formula = - Σ [ y log(p) + (1 - y) log(1 - p) ]

```python
def bce(actual, predicted):
  s = np.sum(actual *np.log(predicted) + (1 - actual) * np.log(1 - predicted))
  return -1 * s

bce(actual, predicted)
```
