### why numpy is faster

 NumPy is faster because:
 - Uses C internally
 - Avoids Python loops (vectorization)
 - Uses contiguous memory
 - Fixed data types

### array operation (scalar operation)

```python
#relational
a1 > 5
```

### array operation

```python
a1 = np.random.random((3, 3))   # 3x3 matrix with values between 0 and 1
a1 = np.round(a1 * 100)         # scale to 0–100 and round
a1
```

### row/col wise operation

###dot product

```python
#first's col has to be == second's row number
# 3 x 4 and 4 x 3 (valid condition for dot product)
#output will be in 3 x 3

a1 = np.arange(12).reshape(3, 4)
a2 = np.arange(12, 24).reshape(4, 3)
np.dot(a1, a2)
```

###log and exponents

```python
np.log(a1)
np.exp(a2)
```
